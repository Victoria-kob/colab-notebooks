```python
df_gapminder_2019 = df_gapminder[df_gapminder["year"] == 2019].reset_index(drop=True)
df_gapminder_2019
```

## Visualize One Quantitative Variable

```python
fig, ax = plt.subplots(figsize=(7, 5))

ax.hist(df_gapminder_2019["lifeExp"].dropna(), bins=20, color="#4C78A8", edgecolor="white")
ax.set_title("Distribution of Life Expectancy (2019)")
ax.set_xlabel("Life Expectancy (years)")
ax.set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Histogram for Life Expectancy
axes[0].hist(df_gapminder_2019["lifeExp"].dropna(), bins=10, color="#4C78A8", edgecolor="white")
axes[0].set_title("Distribution of Life Expectancy (2019)")
axes[0].set_xlabel("Life Expectancy (years)")
axes[0].set_ylabel("Count")

# Histogram for GDP per Capita
axes[1].hist(df_gapminder_2019["gdpPercap"].dropna(), bins=10, color="#F58518", edgecolor="white")
axes[1].set_title("Distribution of GDP per Capita (2019)")
axes[1].set_xlabel("GDP per Capita (thousand USD)")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Density plot for Life Expectancy
df_gapminder_2019["lifeExp"].dropna().plot(kind="density", ax=axes[0], color="#4C78A8", linewidth=2)
axes[0].set_title("Density Plot of Life Expectancy (2019)")
axes[0].set_xlabel("Life Expectancy (years)")
axes[0].set_ylabel("Density")

# Density plot for GDP per Capita
df_gapminder_2019["gdpPercap"].dropna().plot(kind="density", ax=axes[1], color="#F58518", linewidth=2)
axes[1].set_title("Density Plot of GDP per Capita (2019)")
axes[1].set_xlabel("GDP per Capita (thousand USD)")
axes[1].set_ylabel("Density")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Life Expectancy - Histogram
axes[0, 0].hist(df_gapminder_2019["lifeExp"].dropna(), bins=20, color="#4C78A8", edgecolor="white", alpha=0.7)
axes[0, 0].set_title("Histogram: Life Expectancy (2019)")
axes[0, 0].set_xlabel("Life Expectancy (years)")
axes[0, 0].set_ylabel("Count")

# Life Expectancy - Density
df_gapminder_2019["lifeExp"].dropna().plot(kind="density", ax=axes[0, 1], color="#4C78A8", linewidth=2)
axes[0, 1].set_title("Density Plot: Life Expectancy (2019)")
axes[0, 1].set_xlabel("Life Expectancy (years)")
axes[0, 1].set_ylabel("Density")

# GDP per Capita - Histogram
axes[1, 0].hist(df_gapminder_2019["gdpPercap"].dropna(), bins=20, color="#F58518", edgecolor="white", alpha=0.7)
axes[1, 0].set_title("Histogram: GDP per Capita (2019)")
axes[1, 0].set_xlabel("GDP per Capita (thousand USD)")
axes[1, 0].set_ylabel("Count")

# GDP per Capita - Density
df_gapminder_2019["gdpPercap"].dropna().plot(kind="density", ax=axes[1, 1], color="#F58518", linewidth=2)
axes[1, 1].set_title("Density Plot: GDP per Capita (2019)")
axes[1, 1].set_xlabel("GDP per Capita (thousand USD)")
axes[1, 1].set_ylabel("Density")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Life Expectancy - Histogram with Density overlay
ax1 = axes[0]
ax1_twin = ax1.twinx()

ax1.hist(df_gapminder_2019["lifeExp"].dropna(), bins=20, color="#4C78A8", edgecolor="white", alpha=0.6)
ax1.set_xlabel("Life Expectancy (years)")
ax1.set_ylabel("Frequency", color="#4C78A8")
ax1.tick_params(axis='y', labelcolor="#4C78A8")

df_gapminder_2019["lifeExp"].dropna().plot(kind="density", ax=ax1_twin, color="#2C5074", linewidth=2)
ax1_twin.set_ylabel("Kernel Density", color="#2C5074")
ax1_twin.tick_params(axis='y', labelcolor="#2C5074")
ax1.set_title("Life Expectancy (2019): Histogram & Density")

# GDP per Capita - Histogram with Density overlay
ax2 = axes[1]
ax2_twin = ax2.twinx()

ax2.hist(df_gapminder_2019["gdpPercap"].dropna(), bins=20, color="#F58518", edgecolor="white", alpha=0.6)
ax2.set_xlabel("GDP per Capita (thousand USD)")
ax2.set_ylabel("Frequency", color="#F58518")
ax2.tick_params(axis='y', labelcolor="#F58518")

df_gapminder_2019["gdpPercap"].dropna().plot(kind="density", ax=ax2_twin, color="#C56510", linewidth=2)
ax2_twin.set_ylabel("Kernel Density", color="#C56510")
ax2_twin.tick_params(axis='y', labelcolor="#C56510")
ax2.set_title("GDP per Capita (2019): Histogram & Density")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
sns.histplot(data = df_gapminder_2019, x = "gdpPercap")
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(df_gapminder_2019["gdpPercap"], df_gapminder_2019["lifeExp"], 
           color="#4C78A8", alpha=0.6, edgecolor="white", s=80)
ax.set_title("Life Expectancy vs GDP per Capita (2019)")
ax.set_xlabel("GDP per Capita (thousand USD)")
ax.set_ylabel("Life Expectancy (years)")
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
sns.jointplot(data=df_gapminder_2019, x="gdpPercap", y="lifeExp", 
              joint_kws={"alpha": 0.6, "s": 80, "edgecolor": "white"},
              height=6, ratio=4)

plt.suptitle("Life Expectancy vs GDP per Capita (2019)", y=1.02)
plt.xlabel("GDP per Capita (thousand USD)")
plt.ylabel("Life Expectancy (years)")

plt.tight_layout()
plt.show()
```

```python
g = sns.jointplot(data=df_gapminder_2019, x="gdpPercap", y="lifeExp", kind="reg",
                  joint_kws={"scatter_kws": {"alpha": 0.5, "edgecolor": "white"}},
                  height=7)

g.set_axis_labels("GDP per Capita (thousand USD)", "Life Expectancy (years)", fontsize=12)
g.figure.suptitle("Life Expectancy vs GDP per Capita (2019)", y=1.02, fontsize=14)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9, 6))

plot_df = df_gapminder_2019[["gdpPercap", "lifeExp"]].dropna().copy()
sizes = (plot_df["gdpPercap"] / plot_df["gdpPercap"].max()) * 1500 + 20

sc = ax.scatter(
    plot_df["gdpPercap"],
    plot_df["lifeExp"],
    s=sizes,
    c=plot_df["lifeExp"],
    cmap="viridis",
    alpha=0.6,
    edgecolor="white",
    linewidth=0.6
)

ax.set_title("Bubble Plot: Life Expectancy vs GDP per Capita (2019)")
ax.set_xlabel("GDP per Capita (thousand USD)")
ax.set_ylabel("Life Expectancy (years)")
ax.grid(True, alpha=0.25)

cbar = plt.colorbar(sc, ax=ax)
cbar.set_label("Life Expectancy (years)")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

plot_df = df_gapminder_2019[["country_name", "gdpPercap", "lifeExp"]].dropna().copy()
sizes = (plot_df["gdpPercap"] / plot_df["gdpPercap"].max()) * 1000 + 30

sc = ax.scatter(
    plot_df["gdpPercap"],
    plot_df["lifeExp"],
    s=sizes,
    c=plot_df["gdpPercap"],
    cmap="coolwarm",
    alpha=0.65,
    edgecolor="white",
    linewidth=0.7
)

# Annotate top 10 countries by gdpPercap
top10 = plot_df.nlargest(10, "gdpPercap")
for _, row in top10.iterrows():
    ax.annotate(
        row["country_name"],
        xy=(row["gdpPercap"], row["lifeExp"]),
        fontsize=7,
        ha="left",
        va="bottom",
        color="black"
    )

ax.set_title("Bubble Chart: Life Expectancy vs GDP per Capita (2019)", fontsize=14)
ax.set_xlabel("GDP per Capita (thousand USD)", fontsize=12)
ax.set_ylabel("Life Expectancy (years)", fontsize=12)
ax.grid(True, alpha=0.25)

cbar = plt.colorbar(sc, ax=ax)
cbar.set_label("GDP per Capita (thousand USD)")

plt.tight_layout()
plt.show()
```
