```python
df_s_china = df[df['country_name'] == 'China'][['country_name', 'year', 'e_gdp', 'e_gdppc']]
df_s_china
```

```python
df_s_china_modern = df_s_china[df_s_china['year'] >= 1949].reset_index(drop=True)
df_s_china_modern
```

```python
import numpy as np

conditions = [
    (df_s_china_modern['year'] >= 1949) & (df_s_china_modern['year'] <= 1976),
    (df_s_china_modern['year'] > 1976) & (df_s_china_modern['year'] <= 1989),
    (df_s_china_modern['year'] > 1989) & (df_s_china_modern['year'] <= 2002),
    (df_s_china_modern['year'] > 2002) & (df_s_china_modern['year'] <= 2013),
    (df_s_china_modern['year'] > 2013) & (df_s_china_modern['year'] <= 2024)
]
choices = ['Mao', 'Hua-Hu-Zhao', 'Jiang', 'Hu', 'Xi']

df_s_china_modern['leadership'] = np.select(conditions, choices, default=None)
df_s_china_modern
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 1, figsize=(12, 6), sharex=True)

for ax, col in zip(axes, ['e_gdp', 'e_gdppc']):
    missing = df_s_china_modern['year'][df_s_china_modern[col].isna()]
    not_missing = df_s_china_modern['year'][df_s_china_modern[col].notna()]
    ax.scatter(not_missing, [1]*len(not_missing), color='green', label='Available', marker='|', s=200)
    ax.scatter(missing, [1]*len(missing), color='red', label='Missing', marker='|', s=200)
    ax.set_ylabel(col)
    ax.set_yticks([])
    ax.legend(loc='upper left')

plt.xlabel('Year')
plt.suptitle('Missing values by year for e_gdp and e_gdppc in China (1949-2024)')
plt.tight_layout()
plt.show()
```

```python
df_s_china_modern = df_s_china_modern[df_s_china_modern['year'] <= 2019].reset_index(drop=True)
df_s_china_modern
```

```python
df_s_china_modern[df_s_china_modern['year'] != 1950].reset_index(drop=True)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 10))

leadership_colors = {
    'Mao': 'red',
    'Hua-Hu-Zhao': 'orange',
    'Jiang': 'green',
    'Hu': 'blue',
    'Xi': 'purple'
}

for ax, col in zip(axes, ['e_gdp', 'e_gdppc']):
    for leadership, group in df_s_china_modern.groupby('leadership'):
        ax.plot(group['year'], group[col], 
                color=leadership_colors[leadership], 
                label=leadership, linewidth=2)
        ax.fill_between(group['year'], group[col], 
                        alpha=0.1, color=leadership_colors[leadership])
    ax.set_ylabel(col, fontsize=12)
    ax.set_xlabel('Year', fontsize=12)
    ax.legend(title='Leadership', loc='upper left')
    ax.grid(True, linestyle='--', alpha=0.5)

plt.suptitle("China's Economic Performance under Different Leadership", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 10))

for ax, col in zip(axes, ['e_gdp', 'e_gdppc']):
    df_growth = df_s_china_modern.copy()
    df_growth[f'{col}_yoy'] = df_growth[col].pct_change() * 100  # Year-on-year % change

    for leadership_name, group in df_growth.groupby('leadership', sort=False):
        ax.bar(group['year'], group[f'{col}_yoy'],
               color=leadership_colors[leadership_name],
               label=leadership_name, alpha=0.8)

    ax.axhline(0, color='black', linewidth=0.8, linestyle='--')
    ax.set_ylabel(f'{col} YoY Growth (%)', fontsize=12)
    ax.set_xlabel('Year', fontsize=12)
    ax.legend(title='Leadership', loc='upper left')
    ax.grid(True, linestyle='--', alpha=0.4)

plt.suptitle("China's Economic Growth Rate (YoY %) under Different Leadership", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
df_s_china_japan = df[df['country_name'].isin(['China', 'Japan'])][['country_name', 'year', 'e_gdp', 'e_gdppc']]
df_s_china_japan = df_s_china_japan[(df_s_china_japan['year'] >= 1949) & (df_s_china_japan['year'] <= 2019)].reset_index(drop=True)

fig, axes = plt.subplots(2, 1, figsize=(14, 10))

country_colors = {'China': 'red', 'Japan': 'blue'}

for ax, col in zip(axes, ['e_gdp', 'e_gdppc']):
    for country, group in df_s_china_japan.groupby('country_name'):
        ax.plot(group['year'], group[col],
                color=country_colors[country],
                label=country, linewidth=2)
        ax.fill_between(group['year'], group[col],
                        alpha=0.1, color=country_colors[country])
    ax.set_ylabel(col, fontsize=12)
    ax.set_xlabel('Year', fontsize=12)
    ax.legend(title='Country', loc='upper left')
    ax.grid(True, linestyle='--', alpha=0.5)

plt.suptitle("Economic Performance: China vs Japan (1949-2019)", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
